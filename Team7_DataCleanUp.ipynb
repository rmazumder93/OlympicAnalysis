```python
import pandas as pd
```

```python
dic = pd.read_csv("dictionary.csv")
summer = pd.read_csv("summer.csv")
winter = pd.read_csv("winter.csv")
```

```python
#dropping all the NAs
dic = dic.dropna()
summer = summer.dropna()
winter = winter.dropna()
```

```python
#creating column to distinguin different types of olympics
summer["Type"] = "Summer"
winter["Type"] = "Winter"
```

```python
#appending summer and winter data into one
olympics = summer.append(winter)
```

```python
#filtering out data to have datasets after 1900 
olympics_100years = olympics[olympics["Year"] > 1900]
```

```python
#renaming columns and merging data with the GDP and dropping duplicates
olympics_100years = olympics_100years.rename(columns={"Country":"Code"})
olympics_gdp = olympics_100years.merge(dic, how="inner", on="Code")
olympics_gdp = olympics_gdp.drop_duplicates()
```

```python
#exporting data
olympics_gdp.to_csv("olympics_data.csv", index=False)
```
